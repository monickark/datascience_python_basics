# CSV

```python
from io import StringIO,BytesIO
```

```python
data = ('col1,col2,col3\n'
        'x,y,1\n'
        'a,b,2\n'
       'c,d,3\n')
```

```python
type(data)
```

```python
pd.read_csv(StringIO(data))
```

```python
df=pd.read_csv(StringIO(data), usecols=['col1', 'col3'])
```

```python
df
```

```python
df.to_csv('Test1.csv')
```

```python
## Specifying column data types
```

```python
data = ('a,b,c,d\n'
        '1,2,3,4\n'
        '4,6,7,8\n'
       '9,10,11\n')
```

```python
df = pd.read_csv(StringIO(data), dtype=object)
```

```python
df
```

```python
df['a']
```

```python
df['a'][1]
```

```python
## setting different datatype for columns
df = pd.read_csv(StringIO(data),dtype={'b':int, 'c':float, 'a':'Int64'}) 
```

```python
df
```

```python
type(df['a'][1])
```

```python
df.dtypes
```

```python
df.dtypes['a']
```

```python
df.dtypes['c']
```

```python
df.dtypes['c']
```

```python
## index columns and training delimiters

data = ('index, a, b, c \n'
        '4, apple, bat, 5.7\n'
           '8, Orange, cow, 10')
```

```python
df = pd.read_csv(StringIO(data))
```

```python
df
```

```python
df = pd.read_csv(StringIO(data), index_col=0)
```

```python
df
```

```python
## Quoting and escape char useful in NLP
```

```python
data = 'a,b\n"hello, \\"Bob\\", nice to see you", 5'
```

```python
df = pd.read_csv(StringIO(data), escapechar='\\')
```

```python
df
```
