### Problem to identify how much tip would be given based on other parameters


```python
import seaborn as sns
df = sns.load_dataset("tips")
```

```python
df.head()
```

tip -> dependent feature
Remaining are independent feature

```python
df.dtypes
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

```python
sns.jointplot(x="tip",y="total_bill", data=df, kind='hex')
```

```python
sns.jointplot(x="tip",y="total_bill", data=df, kind='reg')
```

## Pair plots

Scatter plots which shows relationship between 2 fields

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue="sex")
```

## Countplot

```python
sns.countplot(data=df)
```

```python
sns.countplot(x='sex', data=df)
```

in barplot we can provide both x and y axis.

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
```
